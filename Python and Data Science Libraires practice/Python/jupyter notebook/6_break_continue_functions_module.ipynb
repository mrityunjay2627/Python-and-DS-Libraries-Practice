# Break Keyword

break keyword us used to break out from a loop

```python
for i in range(9):
    if i >3:
        break
    print(i)
    
print('termination of for loop')
```

```python
i = 0
while i < 9:
    print(i)
    if i==3:
        break
    i += 1
```

```python
# recusrive function

'''- function calling itself

- factorial of a number'''

def fact(num):
    if num == 0:
        return 1
    else:
        return num * fact(num-1)
```

```python
fact(6)
```
