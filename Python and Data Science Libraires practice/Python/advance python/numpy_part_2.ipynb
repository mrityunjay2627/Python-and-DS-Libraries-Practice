```python
# vector stacking  /concatination

a = np.array([["New York","Maryland"],["Albany","Annapolis"]])

b = np.array([["Ohio","Arizona",],["Columbus","Phoenix"]])
```

```python
# axis = 0 - rowise - make sure column is same for both the array
np.concatenate((a,b))
```

```python
# axis = 1 - column - make sure rows is same for both the array
np.concatenate((a,b), axis = 1)
```

```python
# copy and view
```

```python
arr = np.array([1,2,3,4,5])
x = arr.copy()

print(arr)
print(x)

x[0] = 100

print(arr)
print(x)
```

```python
arr = np.array([1,2,3,4,5])
x = arr.view()

print(arr)
print(x)

x[0] = 100

print(arr)
print(x)
```

```python
arr = np.array([1,2,3,4,5])
x = arr

print(arr)
print(x)

x[0] = 100

print(arr)
print(x)
```
