```python
# slicing operations

- slicing operation always give you a new dataframe

1. normal way
2. loc
3. iloc
```

```python
#1. Normal slicing - it does the slicing only on the rows

syntax - dataframe[start:stop:step]  - zero indexing

or

syntax - dataframe[row_start_name : row_stop_name : step]
```

```python
bike_share
```

```python
bike_share[0:5]
```

```python
bike_share[0:20:2]
```

```python
new_bike_data = bike_share[0:6]
```

```python
new_bike_data
```

```python
custom_index = ['row_1' , 'row_2' , 'row_3' , 'row_4' , 'row_5' , 'row_6']

new_bike_data.index = custom_index
```

```python
new_bike_data
```

```python
# zero indexing
new_bike_data[0:3]
```

```python
new_bike_data[new_bike_data['usertype'] == 'Subscriber']
```

```python
# custom indexing
new_bike_data['row_1':'row_3']
```

```python
2. loc method

- does the slicing based on the rows as well as column also
syntax - df.loc[row_index_start:row_index_stop:step , column_start : column_stop : step]
```

```python
3 . iloc 

- also does the slicing based on row and columns

syntax - df.iloc[start_row : stop_row  : step , start:stop:step]
```

```python
new_bike_data
```

```python
new_bike_data.iloc[0:3 , 0:4]
```

```python
new_bike_data.iloc[new_bike_data['usertype'] == 'Subscriber']
```

```python
new_bike_data.iloc[[0,2,3,4] , 0:4]
```

```python
1 .normal - only row can be slicing
filtering can be done only on rows
index and custom

2. loc - both rows and column
- filtering on both
- custom indexing


3. iloc - bot rows and columns
- zero indexing
```

```python
# usertype is subscriber and bikeid is less than 30000

new_bike_data.loc['usertype'=='subscriber' , 'bikeid'<3000] 
```

```python
new_bike_data.loc[(new_bike_data['usertype'] == 'Subscriber')][new_bike_data['bikeid'] < 30000] 
```

```python
new_bike_data.loc[(new_bike_data['usertype'] == 'Subscriber') & 
               (new_bike_data['bikeid'] < 30000) ]
```

```python
new_bike_data.loc[(new_bike_data['usertype'] == 'Subscriber') | 
               (new_bike_data['bikeid'] < 30000) ]
```

```python
# iterate over dataframes

1. itertuples
2. iterrows
3. iteritems
```

```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
                   'Headquarters' : ["Paris", "Bangalore", "Washington",
                                     "Tokyo", "Beijing"],
                   'Country_name' : ["France", "India", "USA", "Japan", "China"],
                 'Formation_year' : [1961, 1969, 1958, 2003, 1993]
             }
```

```python
df = pd.DataFrame(space_org)
```

```python
df
```

```python
for i in df.itertuples():
    print(i)
```

```python
for i in df.itertuples():
    if i.Headquarters == 'Bangalore':
        print(i.Organizations_name)
```

```python
for i in df.iterrows():
    print(i , end = '\n\n')
```

```python
df
```

```python
for i in df.iteritems():
    print(i)
```

```python
# merge
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
pd.merge(df1 , df2 , how = 'outer' , on='city')
```

```python
pd.merge(df1 , df2 , how = 'inner' , on='city')
```

```python
pd.merge(df1 , df2 , how = 'left' , on='city')
```

```python
pd.merge(df1 , df2 , how = 'right' , on='city')
```

# handling missing data

```python
# fillna

df = pd.read_csv('datasets/weather_data_nan.csv')
```

```python
df
```

```python
df.fillna(0)
```

```python
df.fillna({
    'temperature' : 0,
    'windspeed': 0,
    'event':'no event'
})
```

```python
df.event.mode()
```

```python
df.fillna({
'temperature': df.temperature.mean(),
'windspeed': df.windspeed.median(),
'event':df.event.mode()[0]
}) 
```

```python
df
```

```python
# dropna

df.dropna(axis = 0)
```

```python
df.dropna(axis = 1)
```
