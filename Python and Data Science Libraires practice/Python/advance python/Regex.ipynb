```python
text_to_search = '''
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ\s
321-555-4321
1234567890
Ha HaHa
MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
khanafsaan11.com
321-555-4321
123.555.1234
123*555*-1234
123.555.1234
800-555-1234
900-555-1234
Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
Mr_hello
'''
```

```python
pattern = re.compile(r'\b\d\d\d\b')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# extract all the phone number that are present (123-345-5678)

pattern = re.compile(r'\b\d\d\d-\d\d\d-\d\d\d\d\b')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# extract all the phone number that are present (123-345-5678 / 123.345.5678)

# [] - matches the character in the bracket

pattern = re.compile(r'\b\d\d\d[-.]\d\d\d[-.]\d\d\d\d\b')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
pattern = re.compile(r'[az0]')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# [^ ] - matches the chara that is not in the bracket

pattern = re.compile(r'[^012345]')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# {3} - exact number
# {3,4} - range of numbers (minimum , Maximum)


pattern = re.compile(r'\b\d{3}[-.]\d{3}[-.]\d{4}\b')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# step 5

pattern = re.compile(r'Mr\. ')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# step 6

pattern = re.compile(r'Mr\. [A-Z]')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# step 7 - 

pattern = re.compile(r'Mr\. [A-Z][a-z]')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# step 8 - *

pattern = re.compile(r'Mr\. [A-Z][a-z]*')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# step 9 - ?

pattern = re.compile(r'Mr\.? [A-Z][a-z]*')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# step 10 - 

pattern = re.compile(r'M[rs]\.? [A-Z][a-z]*')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```
