```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.relplot(x = 'total_bill' , y = 'tip' , data = tips)
```

```python
"""
hue - categrocial column
hue - categoirizing based on color

"""

sns.relplot(x = 'total_bill' , y = 'tip' , data = tips , hue = 'smoker')
```

```python
"""
style - categrocial column
style - categoirizing based on different styles

"""

sns.relplot(x = 'total_bill' , y = 'tip' , data = tips , hue = 'smoker' , style = 'time')
```

```python
"""
style - categrocial column
style - categoirizing based on different styles

"""

sns.relplot(x = 'total_bill' , y = 'tip' , data = tips , hue = 'smoker' , style = 'time' , size = 'size')
```

```python
sns.relplot(x = 'total_bill' , y = 'tip' , data = tips , col = 'smoker')
```
